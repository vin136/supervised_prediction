```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

from pprint import pprint
from typing import Optional, Tuple

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy
import seaborn as sns
import tensorflow as tf
from tensorflow.python import tf2
import tensorflow_probability as tfp
#import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

sns.reset_defaults()
sns.set_context(context='talk',font_scale=0.7)

%matplotlib inline

tfd = tfp.distributions
tf.get_logger().setLevel('ERROR')
import numpy as np
```

```python
# Normalized st. deviation
def std_normalized(vals):
    return np.std(vals) / np.mean(vals)

# Ratio of diff between last price and mean value to last price
def ma_ratio(vals):
    return (vals[-1] - np.mean(vals)) / vals[-1]

# z-score for volumes and price
def values_deviation(vals):
    return (vals[-1] - np.mean(vals)) / np.std(vals)
```

```python

start_date_string = '2014-04-01'
asset = 'BANKBARODA'
column_price = 'Adj Close'
column_high = 'High'
column_low = 'Low'
column_volume = 'Volume'

import yfinance
data = yfinance.download("BANKBARODA.NS", start=start_date_string,end="2021-03-22")
data.dropna(inplace=True)
dataset = data.copy()
dataset = dataset.shift(1)
print(dataset.columns)
```

```python
plt.figure(figsize=(20,10))
plt.plot(dataset[column_volume])
plt.title(asset)
plt.show()
```

```python
# Feature params
future_period = 1
std_period = 10
ma_period = 10
price_deviation_period = 10
volume_deviation_period = 10

# Create features
cols_features = ['last_return', 'std_normalized', 'ma_ratio', 'price_deviation', 'volume_deviation']
labels = ['future_return']
dataset['last_return'] = dataset[column_price].pct_change()
dataset['std_normalized'] = dataset[column_price].rolling(std_period).apply(std_normalized)
dataset['ma_ratio'] = dataset[column_price].rolling(ma_period).apply(ma_ratio)
dataset['price_deviation'] = dataset[column_price].rolling(price_deviation_period).apply(values_deviation)
dataset['volume_deviation'] = dataset[column_volume].rolling(volume_deviation_period).apply(values_deviation)

```

```python
#output
dataset["future_return"] = dataset[column_price].pct_change(future_period).shift(-future_period)
```

```python
dataset = dataset.replace([np.inf, -np.inf], np.nan)
dataset = dataset.dropna()
```

```python
def get_train_valid_test(df,split=[0.6,0.5],seed=0):
    np.random.seed(seed)
    train_end_id = int(len(df)*split[0])
    train_dset = df.iloc[:train_end_id]
    rem_rows = np.arange(train_end_id,len(df))
    np.random.shuffle(rem_rows)
    vld_end_id = int(len(rem_rows)*split[-1])
    valid_dset = df.iloc[rem_rows[:vld_end_id]]
    test_dset= df.iloc[rem_rows[vld_end_id:]]
    return train_dset,valid_dset,test_dset
```

```python
trn,vld,tst = get_train_valid_test(dataset)
```

```python
# Some utilities

def plot_history(history: tf.keras.callbacks.History) -> None:
    """
    Plot training vs validation MAE and MSE over epoch.
    """
    hist = pd.DataFrame(history.history)
    hist['epoch'] = history.epoch

    plt.figure()
    plt.xlabel('Epoch')
    plt.ylabel('Mean Abs Error [MPG]')
    plt.plot(hist['epoch'], hist['mae'],
           label='Train Error')
    plt.plot(hist['epoch'], hist['val_mae'],
           label = 'Val Error')
    plt.ylim([0,0.1])
    plt.legend()

    plt.figure()
    plt.xlabel('Epoch')
    plt.ylabel('Mean Square Error [$MPG^2$]')
    plt.plot(hist['epoch'], hist['mse'],
           label='Train Error')
    plt.plot(hist['epoch'], hist['val_mse'],
           label = 'Val Error')
    plt.ylim([0,0.02])
    plt.legend()
    plt.show()


```

```python
def build_model() -> tf.keras.Model:
    model = keras.Sequential([
        layers.Dense(32, activation='relu', input_shape=[len(x_trn_std.keys())]),
        layers.Dense(32, activation='relu'),
        layers.Dense(3,activation='softmax'),
    ])


    model.compile(loss='sparse_categorical_crossentropy',
                optimizer='adam',
                metrics=['mae', 'mse'])
    return model
```

```python
model = build_model()
model.summary()
```

```python
class PrintDot(keras.callbacks.Callback):
    """
    Display training progress by printing a single dot for 
    each completed epoch.
    """
    def on_epoch_end(self, epoch, logs) -> None:
        """
        `on_epoc_end` hook to take the epoch number and logs.
        This will run whenever the training loop finishes an epoch.
        """
        if epoch % 100 == 0: 
            print('')
        print('.', end='')

        
EPOCHS = 1000


history = model.fit(
    x_trn_std.values, 
    y_trn.values,
    epochs=EPOCHS, 
    validation_data = (x_vld_std.values,y_vld.values), 
    verbose=0,
    callbacks=[PrintDot()]
)
```

```python
def build_dropout_model() -> tf.keras.Model:
    model = keras.Sequential([
        layers.Dense(10, activation='relu',  input_shape=[len(x_trn_std.keys())]),
        layers.Dense(5, activation='relu'),
        layers.Dense(1,activation='tanh'),
    ])

    optimizer = tf.keras.optimizers.RMSprop(0.001)

    model.compile(loss='mse',
                optimizer=optimizer,
                metrics=['mae', 'mse'])
    return model

dropout_model = build_dropout_model()

history_dropout = dropout_model.fit(
    x_trn_std.values, 
    y_trn.values,
    epochs=EPOCHS, 
    validation_data = (x_vld_std.values,y_vld.values), 
    verbose=0,
    callbacks=[PrintDot()]
)
plot_history(history_dropout)
```

# Modeling with probability distributions

```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

from pprint import pprint
from typing import Optional, Tuple

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy
import seaborn as sns
import tensorflow as tf
from tensorflow.python import tf2
import tensorflow_probability as tfp


sns.reset_defaults()
sns.set_context(context='talk',font_scale=0.7)

%matplotlib inline

tfd = tfp.distributions
tf.get_logger().setLevel('ERROR')

model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu',  input_shape=[len(x_trn_std.keys())]),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(1),
    tfp.layers.DistributionLambda(
            lambda t: tfd.Normal(
                loc=t,
                scale=0.03)
        ),
])
```

```python
model.compile(optimizer=tf.optimizers.RMSprop(0.01), loss='mse')
history_dropout_dst = model.fit(
    x_trn_std.values, 
    y_trn.values,
    epochs=EPOCHS, 
    validation_data = (x_vld_std.values,y_vld.values), 
    verbose=0,
    callbacks=[PrintDot()]
)
```

```python
negloglik = lambda y, rv_y: -rv_y.log_prob(y)
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu',  input_shape=[len(x_trn_std.keys())]),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(2),
    tfp.layers.DistributionLambda(
            lambda t: tfd.Normal(
                loc=t[..., :1],
                scale=1e-3 + tf.math.softplus(0.05 * t[..., 1:])  # Fit the standard deviation to the second term
            )
        ),
])
```

```python

# Due to TensorFlow 2.0 Eager semantics, we can use `negloglik` directly as our loss function.
model.compile(optimizer=tf.optimizers.Adam(learning_rate=0.01), loss=negloglik)
```

```python
history_dropout_dst2 = model.fit(
    x_trn_std.values, 
    y_trn.values,
    epochs=EPOCHS, 
    validation_data = (x_vld_std.values,y_vld.values), 
    verbose=0,
    callbacks=[PrintDot()]
)
```
